```python
import pandas as pd
from pathlib import Path
```

```python
# Import data
feb20_path = Path("../Histories/DAT_XLSX_EURUSD_M1_202002.csv")
mar20_path = Path("../Histories/DAT_XLSX_EURUSD_M1_202003.csv")

# Read in data and index by DateTime
feb20_forexdata = pd.read_csv(feb20_path, index_col='DateTime')
mar20_forexdata = pd.read_csv(mar20_path, index_col='DateTime')
```

```python
#Show February 2020 Data
feb20_forexdata.head()
```

```python
#Show February 2020 Data
mar20_forexdata.head()
```

```python
 # Join datasets by row
joined_forexdata = pd.concat([feb20_forexdata, mar20_forexdata], axis="rows", join="inner")
joined_forexdata
```

```python
joined_forexdata.to_csv("../Histories/joined_forexdata.csv")
```
