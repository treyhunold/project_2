```python
sma_df = fd.loc[:, ['Close']].copy()
```

```python
sma_df.head()
```

```python
def sma(df):
    ''' This function creates a simple moving averagewith windows of 5 and 15 appends to new columns, "signals and Entry/Exit"
    '''
    short_window = 5
    long_window = 15
    
    df['SMA5'] = df['Close'].rolling(window=short_window).mean()
    df['SMA15'] = df['Close'].rolling(window=long_window).mean()
    
    df['Signal'] = 0.0
    df["Signal"][short_window:] = np.where(
    df["SMA5"][short_window:] < df["SMA15"][short_window:], 1.0, 0.0)
    
    df['Entry/Exit'] = df['Signal'].diff()

    fd = df

    return fd
```

```python
fd = sma(sma_df)
```

```python
fd.head()
```

```python
fd.dtypes
```

```python
fd = fd.astype({"Close": object, "SMA5": object,
               'SMA15': object, "Signal": object,
               "Entry/Exit": object})
fd.dtypes
```

```python
# Visualize exit position relative to close price

exit = fd[fd['Entry/Exit'] == -1.0]['Close'].hvplot.scatter(color='red',legend=False,ylabel='Price in EURO/DOL',width=1000, height=400)

# Visualize entry position relative to close price
entry = fd[fd['Entry/Exit'] == 1.0]['Close'].hvplot.scatter(color='green',legend=False,ylabel='Price in EURO/DOL', width=1000, height=400)

# Visualize close price for the investment
security_close = fd[['Close']].hvplot(line_color='lightgray',ylabel='Price in EURO/DOL',width=1000,height=400)

# Visualize moving averages
moving_avgs = fd[['SMA50', 'SMA100']].hvplot(ylabel='Price in EURO/DOL',width=1000,height=400)

# Overlay plots
entry_exit_plot = security_close * moving_avgs * entry * exit
entry_exit_plot.opts(xaxis=None)
```
